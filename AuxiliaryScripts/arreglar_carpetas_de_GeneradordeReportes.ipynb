```python
import os
import shutil

# Cambia esta ruta a la carpeta raíz de tus outputs
BASE_DIR = r"C:\Users\HeyCe\World Tree Technologies Inc\Operations - Documentos\WorldTreeSystem\GeneradordeReportes\outputs"

for root, dirs, files in os.walk(BASE_DIR):
    if "Resumen" in dirs:
        resumen_path = os.path.join(root, "Resumen")
        parent_path = root  # carpeta padre a donde moverás

        print(f"Procesando: {resumen_path}")
        for fname in os.listdir(resumen_path):
            src = os.path.join(resumen_path, fname)
            dst = os.path.join(parent_path, fname)
            # Si el archivo ya existe arriba, elimina solo el de Resumen
            if os.path.exists(dst):
                print(f"Archivo ya existe arriba, eliminando {src}")
                os.remove(src)
            else:
                print(f"Moviendo {src} -> {dst}")
                shutil.move(src, dst)
        # Cuando esté vacía, borra la carpeta Resumen
        try:
            os.rmdir(resumen_path)
            print(f"Borrada carpeta vacía: {resumen_path}")
        except OSError as e:
            print(f"No se pudo borrar {resumen_path}: {e}")
```
