```python
from core.libs import pd
from core.db import get_engine

engine = get_engine()
```

```python
def print_headers(schema, table):
    q = text(f"""
        SELECT column_name, data_type
        FROM information_schema.columns
        WHERE table_schema = '{schema}'
          AND table_name = '{table}'
        ORDER BY ordinal_position;
    """)
    df = pd.read_sql(q, engine)
    print(f"\n=== {schema}.{table} ===")
    print(df)

# print_headers("masterdatabase", "contract_tree_information")
# print_headers("masterdatabase", "contract_allocation")
# print_headers("masterdatabase", "farmer_personal_information")
# print_headers("masterdatabase", "inventory_metrics_current")
```
